# Book Store Scraping

```python
from bs4 import BeautifulSoup
import csv 
import requests
```

```python
page=f"https://books.toscrape.com/catalogue/category/books_1/page-1.html"
src=requests.get(page)
```

```python
soup=BeautifulSoup(src.content,'lxml')
print(soup.prettify())
```

```python
#extracts the books information
books_details=[]
books=soup.find_all('article',{'class':'product_pod'})
```

```python
# book_title=books[0].find('h3').find('a').attrs['title']
# book_price=books[0].find('div',class_="product_price").find('p').text.strip()
# rating='No star-rating' if 'star-rating' not in str(books[0]) else books[0].find('p',class_="star-rating")['class'][-1]
```

```python
def get_book_info(books):
    for book in books:
        book_title = book.find('h3').find('a').attrs['title']
        book_price = book.find('div', class_="product_price").find('p').text.strip()
        #rating = 'No rating' if 'star-rating' not in str(book) else book.find('p', class_='star-rating')['class'][-1]
        rating='No rating' if 'star-rating' not in str(book) else book.find('p',class_='star-rating')['class'][-1]
#         print(book_title, "\n", book_price, "\n", rating)
#         print("-" * 8)
       
        books_details.append({'book_name':book_title,'book_price':book_price,'rating':rating})
                 
```

```python
page_number=1
while True:
    src=requests.get(f"https://books.toscrape.com/catalogue/category/books_1/page-{page_number}.html")
    if page_number==4:#src.status_code != 200:
        break
    soup=BeautifulSoup(src.content,'lxml')
    #extracts the books information
    books=soup.find_all('article',{'class':'product_pod'})
    get_book_info(books)
    page_number+=1 
headers=books_details[0].keys()    
with open("D:/BookStore.csv",'w',encoding='utf-8') as output:
    writer=csv.DictWriter(output,fieldnames=headers)
    writer.writeheader()
    for book in books_details:
        writer.writerow(book)  
    print("File created!!")
    
```
