```python
def replace_values():
    result=[]
    for i in df.Seats:
        if i==' 2 Seats':
            result.append('2')
        elif i==' 4 Seats':
            result.append('4')
        elif i==' 5 Seats':
            result.append('5')
        elif i==' 6 Seats':
            result.append('6')
        elif i==' 7 Seats':
             result.append('7')
        elif i==' 8 Seats':
            result.append('8')
        elif i==' 9 Seats':
            result.append('9')
        elif i==' 11 Seats':
            result.append('11')  
        elif i==' 12 Seats':
            result.append('12') 
        elif i==' 14 Seats':
            result.append('14') 
        elif i==' 15 Seats':
            result.append('15')
        elif i==' 22 Seats':
            result.append('22') 
        else:
            result.append(np.nan) 
            
    return result        
```

### how to detect and remove outliers using pandas

```python
data = {'Value': [10, 15, 12, 18, 20, 25, 8, 22, 30, 40, 100]}
df = pd.DataFrame(data)
df
```

```python
# Calculate the IQR (Interquartile Range)
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1
Q1,Q3
```

```python
lower=(Q1-1.5*IQR)
upper=(Q3+1.5*IQR)
lower,upper
```

```python
outliers=df[(df['Value']<lower)|(df['Value']>upper)]
outliers
```

```python
no_outliers=df[(df['Value']>=lower)&(df['Value']<=upper)]
no_outliers
```
